### Self Employed
> Are you self-employed?

Since our scope is to provide a model for companies to be aware of their employee's mental health, we have decided to remove those who answered `Yes` in the feature `self_employed`, since that person would be an employer.


```python
df["self_employed"].value_counts()
```

```python
df = df.drop(df[df["self_employed"] == "Yes"].index)
df["self_employed"].value_counts()
```

```python
df = df.drop(df[(df["no_employees"] == "1-5") & (df["self_employed"].isnull())].index)
df["self_employed"].isnull().sum()
```

Above the missing values from self employed and with a company size of 1-5 is dropped, since they might be self-employed

```python
df["self_employed"] = df["self_employed"].fillna("No")
```

The rest of the missing values are filled with "No", since they work in a larger company

```python
df.loc[df["self_employed"].isnull()] 
```

### Work Interfere

> *If you have a mental health condition, do you feel that it interferes with your work?*

Work interfere has 264 missing values. Due to the phrasing of the question, we interpret, that the missing values reflect those who does not identify themselves as having a mental health condition. However, here we see a possibility for error, since the missing values could have been due to a mistake of not answering.

Nontheless, we fill the missing values with `N/A` for `Not Applicable`.

Another inference is that the value `Never` might include participants who did not realise skipping was an option and therefore do not have a mental health condition. However, we interpret that these participants do have a mental health condition, but it simply does not interfere with their work.

```python
df['work_interfere'].unique()
```

```python
df["work_interfere"] = df["work_interfere"].fillna("N/A")
```

```python
print("Missing values after cleaning:", df['work_interfere'].isnull().sum())
print("New categories:", df['work_interfere'].unique())
print("Visual overview of new categories", df["work_interfere"].value_counts().plot.bar())
```

### Age
>	The age of the individual

Here we are checking what's in the feature "age"

```python
df['age'].describe()
```

We see `age` min is equal to -1.7 and `age` max is 1.000000e+11. That is rather odd, seeing as it is impossible for people to be that age. To further inspect, we will check our data's outliers

```python
sns.boxplot(x=df["age"])
```

The above boxplot shows that there is a very high right-skewed distribution. What this tells us, is that there are a lot of high outliers in the `age` column, which is supported by the above discovered max value, which for the sake of my further analysis will be dealt with. Taking an inspection from a "normal" working age, which in our opinion is 18-75, we will inspect the amount of participants that have answered below and above these ages:


https://developerpitstop.com/when-do-software-engineers-retire/

```python
def inspect_age (dataframe, column):
    return dataframe[(dataframe[column]<18) | (dataframe[column]>75)]

inspect_age(df, 'age')
```

We can here see that the outliers are represented by age values that are well above or below the "normal" working age. Seeing as there are only 8 participants that are placed at the ages below 18 and above 75, and therefore these will have no statistical impact, we will remove these.

```python
df = df[(df.age >= 18) & (df.age <= 65)]
print('After removing outliers: ')
print(df.age.describe().astype(int))
sns.boxplot(x=df['age'])
```

Now we have a new and more realistic distribution, since we have removed ages under 18 and above 75, due to our assumption that these ages would not be applicable to a survey for the job market.

```python
print("Our new dataframe is therefore the following size:", df.shape)
df
```
